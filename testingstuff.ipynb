```python
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
```

```python
def load_model():
    model = tf.keras.models.model_from_json(open("model.json", "r").read())
    model.load_weights("weights.h5")
    return model
```

```python
model = load_model()
```

```python
images = []
labels = []
```

```python
!pip install opencv-python
from IPython.display import display, Image
import os
import cv2 as cv
directory = r"garbage_classification"
for category in os.listdir(directory):
    path = os.path.join(directory, category)
    for img in os.listdir(path)[:5]:
        image = cv.imread(os.path.join(path, img))
        image = cv.resize(image, (100, 100));
        images.append(image)
        labels.append(category)
        # display(Image(data=cv.imencode('.jpg', image)[1].tobytes()))
```

```python
display(Image(data=cv.imencode('.jpg', images[30])[1].tobytes()))
print(labels[30])
```

```python
predictions = []

for image in images[:30]:
    prediction = model.predict(np.expand_dims(image, axis=0))
    predictions.append(np.argmax(prediction))
```

```python
for i in range(30):
    print(predictions[i])
    print(labels[i])
    display(Image(data=cv.imencode('.jpg', images[i])[1].tobytes()))
```
